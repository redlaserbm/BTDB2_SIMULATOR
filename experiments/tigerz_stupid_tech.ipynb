```python
import b2sim.engine as b2
import b2sim.analysis as dd
```

```python
rounds = b2.Rounds(0.2)

eco_queue = [
    b2.ecoSend(time = rounds.getTimeFromRound(22), send_name='Zero')
]

farms = [
    b2.initFarm(rounds.getTimeFromRound(19), upgrades = [2,4,0]),
]
buy_queue = [
    [b2.repeatedlyBuySupplyDrops()],
    # [b2.activateIMF(index=0)]
]

initial_state_game = {
    'Cash': 10000,
    'Eco': 3100,
    'Rounds': rounds,
    'Game Round': 23,
    'Farms': farms,
    'Buy Queue': buy_queue,
    'Eco Queue': eco_queue,
    'Supply Drops': {0: 0, 'Elite Sniper Index': 0}
}

game_state = b2.GameState(initial_state_game)
game_state.fastForward(target_round = 26)
dd.viewHistory(game_state)
print("The current cash and eco are (%s,%s)"%(game_state.cash, game_state.eco))
print("We currently have %s supply drops!"%(game_state.supply_drops))
b2.writeLog(game_state.logs, filename = 'tigerz_IMF')
```
